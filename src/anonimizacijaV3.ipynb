```python
from matplotlib import pyplot as plt
# Izberemo signal za ID "38222"
signal_id = "38222"
signal_data = kopija_original_signal_id_st[['time_s', signal_id]]

# ƒåasi iz metapodatkov za ID "38222" (prvi vrstica v metapodatkih)
metadata_times = metadata_podatki.loc[metadata_podatki['Unnamed: 0'] == int(signal_id), ['t1', 't2', 't3', 't4']].values.flatten()
# Ustvarimo graf
plt.figure(figsize=(10,6))
plt.plot(signal_data['time_s'], signal_data[signal_id], label=f'Signal {signal_id}', color='blue')

# Dodamo vertikalne ƒçrte za ƒçase iz metapodatkov
for t in metadata_times:
    plt.axvline(x=t, color='red', linestyle='--', label=f'Time {t} ms')

# Nastavitve grafa
plt.title(f'Signal {signal_id} with Metadata Times')
plt.xlabel('Time (s)')
plt.ylabel('Signal Value')
plt.legend()
plt.grid()
plt.show()
```

```python
result = kopija_original_signal_id_st.copy()
for signal_id in kopija_original_signal_id_st.columns[1:]:
    # Izberemo signal za ID "38222"
    signal_id = signal_id
    signal_data = kopija_original_signal_id_st[['time_s', signal_id]]

    # ƒåasi iz metapodatkov za ID "38222"
    metadata_times = metadata_podatki.loc[metadata_podatki['Unnamed: 0'] == int(signal_id), ['t1', 't2', 't3', 't4']].values.flatten()

    # Pretvorimo ƒçase v sekunde
    metadata_times_sec = metadata_times

    # Filtriramo signal med t1 in t2 ter med t3 in t4
    filtered_signal_t1_t2 = signal_data[(signal_data['time_s'] >= metadata_times_sec[0]) & (signal_data['time_s'] <= metadata_times_sec[1])]
    filtered_signal_t3_t4 = signal_data[(signal_data['time_s'] >= metadata_times_sec[2]) & (signal_data['time_s'] <= metadata_times_sec[3])]

    # Zdru≈æimo filtrirane podatke
    filtered_signal_data = pd.concat([filtered_signal_t1_t2, filtered_signal_t3_t4])
    result[signal_id] = filtered_signal_data[signal_id]

```
